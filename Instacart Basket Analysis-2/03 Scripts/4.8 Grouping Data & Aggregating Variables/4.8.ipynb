Import Data

```python
# import of ords_prods_merge
df_ords_prods_merge = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'orders_products_merged_new C.pkl'))
```

```python
#first one million rows.
df_ords_prods_merge = df_ords_prods_merge[:1000000]
```

Groupping Data

```python
# group product_name
df_ords_prods_merge.groupby('product_name')
```

Performing single aggregates


```python
df_ords_prods_merge.groupby('department_id').agg({'order_number': ['mean']})
```

Aggregations that can be conducted without use of the agg()

```python
df_ords_prods_merge.groupby('department_id')['order_number'].mean()
```

Performing Multiple Aggregations

```python
# multiple aggregation with multiple order_number
df_ords_prods_merge.groupby('department_id').agg({'order_number': ['mean', 'min', 'max']})
```

create a loyality flag using loc

```python
df_ords_prods_merge['max_order'] = df_ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
df_ords_prods_merge.head(100)
```

Deriving Column with loc

```python
df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
df_ords_prods_merge.loc[(df_ords_prods_merge['max_order'] <= 40) & (df_ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```
