
1a. Create “price_range” column.


```python
# define a function price_label

def price_label(row):
    if pd.isna(row['prices']):
        return 'Not enough data'
    elif row['prices'] <= 5:
        return 'Low-range product'
    elif 5 < row['prices'] <= 15:
        return 'Mid-range product'
    elif row['prices'] > 15:
        return 'High-range product'
```

```python
# Creating a column for Price Range 
df_ords_prods_merge['price_range'] = df_ords_prods_merge.apply(price_label, axis=1)
```

2 Update "busiest day" column to "busiest days"


```python
#Count the orders per day of the week
orders_per_day = df_ords_prods_merge['orders_day_of_week'].value_counts().sort_values(ascending=False)
```

```python
# Top 2 busiest days
busiest_days = orders_per_day.index[:2].tolist()  
# Bottom 2 slowest days
slowest_days = orders_per_day.index[-2:].tolist()  
```

```python
# runing the new For-loop "orders day of the week" 
result = []

for value in df_ords_prods_merge["orders_day_of_week"]:
    if value in busiest_days:
        result.append("Busiest days")
    else:
        result.append("Regularly busy")
```

4 Create new column "busiest_period_of_day" to identify time periods “Most orders,” “Average orders,” and “Fewest orders.”


```python
#  Count the number of orders per hour of the day
orders_per_hour = df_ords_prods_merge['order_time_of_day'].value_counts().sort_values(ascending=False)
```

```python
#Identify thresholds for "Most orders," "Average orders," and "Fewest orders"
# Split the hours into three groups based on the order frequency
most_orders_threshold = orders_per_hour.quantile(0.66)
fewest_orders_threshold = orders_per_hour.quantile(0.33)
```

```python
#Create a function to label the periods of the day
def busiest_period_label(hour):
    if orders_per_hour[hour] >= most_orders_threshold:
        return 'Most orders'
    elif orders_per_hour[hour] <= fewest_orders_threshold:
        return 'Fewest orders'
    else:
        return 'Average orders'
```

```python
#Apply the function to create a new column
df_ords_prods_merge['busiest_period_of_day'] = df_ords_prods_merge['order_time_of_day'].apply(busiest_period_label)
```
